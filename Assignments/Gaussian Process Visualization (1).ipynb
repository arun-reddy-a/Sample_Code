```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# Define the kernel function
kernel = RBF(length_scale=1.0)

# Define the input domain
X = np.linspace(-5, 5, 100).reshape(-1, 1)

# Initialize the GP with the RBF kernel and no mean function
gp = GaussianProcessRegressor(kernel=kernel, alpha=0, normalize_y=False)

# Draw samples from the GP prior
y_samples = gp.sample_y(X, n_samples=5)

# Plot the samples
plt.figure()
for i in range(5):
    plt.plot(X, y_samples[:,i], label=f'Sample {i+1}')
plt.xlabel('input values')
plt.ylabel('output values')
plt.title('Random Samples from GP Prior of RBF Kernel')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the kernel function
def rbf_kernel(x1, x2, length_scale):
    dist = np.sum((x1 - x2) ** 2)
    return np.exp(-0.5 * dist / length_scale ** 2)

# Define the input domain
X = np.linspace(-5, 5, 100).reshape(-1, 1)

# Compute the covariance matrix
n = X.shape[0]
K = np.zeros((n, n))
length_scale = 1.0
for i in range(n):
    for j in range(n):
        K[i,j] = rbf_kernel(X[i], X[j], length_scale)

# Sample from the prior
mean = np.zeros(n)
L = np.linalg.cholesky(K + 1e-8 * np.eye(n))
y_samples = mean.reshape(-1, 1) + L @ np.random.randn(n, 5)

# Plot the samples
plt.figure()
for i in range(5):
    plt.plot(X, y_samples[:,i], label=f'Sample {i+1}')
plt.xlabel('input values')
plt.ylabel('output values')
plt.title('Random Samples from GP Prior of RBF Kernel')
plt.legend()
plt.show()
```
